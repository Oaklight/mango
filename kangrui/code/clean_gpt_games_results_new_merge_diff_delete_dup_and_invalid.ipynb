```python
def get_collection_and_drop_rf(all2all,all_pairs,final_dir_path,task_type):
    assert game_name in final_dir_path
    
    if task_type=='stepnav':
        id_set=set([value['id'] for value in all2all])
    else:
        id_set=set([value['id'] for value in all_pairs])

   
    drop_list=[]
    collection={}
    for idx,file_name in enumerate(os.listdir(final_dir_path)):
        
        final_file_path=osp.join(final_dir_path,file_name)
        with open(final_file_path,'r') as f:
            result_json = json.load(f)
        
        task_id=result_json['task_id']

        if task_id in id_set:
            if task_id in collection.keys():
                collection[task_id].append(final_file_path)
            else:
                collection[task_id]=[final_file_path]
        else:
            drop_list.append(final_file_path)
    
    collection_dup={}
    for k,v in collection.items():
        if len(v)>1:
            collection_dup[k]=v
    return drop_list,collection_dup

```

```python
# clean data for pathgen

task_type="pathgen"
import os
cnt=0
for idx,game_name in enumerate(os.listdir(save_path)):
    print(f'checking {game_name} ...')
    G,actions,locations,all2all,all_pairs,walkthrough=get_game_info(map_dir,game_name)
    
    # iterate pathgen in result-clean
    file_dir=osp.join(save_path,game_name,'results')
    
#     print(file_dir)
    # iterate pathgen for gpt3 and gpt4
    for idx,dir_path in enumerate(os.listdir(file_dir)):
        if not dir_path.startswith(task_type):
            continue
            
        final_dir_path=osp.join(file_dir,dir_path)
        
        drop_list,collection_dup=get_collection_and_drop_rf(all2all,all_pairs,final_dir_path,task_type)
        if len(collection_dup.keys())>0:
            print(game_name,len(collection_dup.keys()))
#         os.sys.exit()
        for file_name in drop_list:
            cnt+=1
            os.remove(file_name)
print(cnt//2)
```
