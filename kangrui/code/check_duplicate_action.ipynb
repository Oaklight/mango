```python
# test if there're duplicate actions for a node
def check_same_action_different_target_node(G, node):
    edges = G.out_edges(node, data=True)
    
    if len(edges) <= 1:
        return False
    
    action_dict={}
    for edge in edges:
#         print(edge)
        action = edge[2]['action']
        target=edge[1]
        if not action in action_dict.keys():
            action_dict[action]=[target]
        else:
            if target!=action_dict[action][0]:
                return True
    return False
            
def check_same_action_different_target_graph(G):
    for node in G.nodes():
        flag=check_same_action_different_target_node(G, node)
        if flag:
            return True
    return False

def find_duplicate_edges(G):
    duplicate_edges = []
    visited_edges = {}

    for edge in G.edges(data=True):
        source, target, data = edge
        action = data.get('action')
        
        key=(source,action)
        if key in visited_edges.keys():
            visited_edges[key].add((source,action,target))
        else:
            visited_edges[key]=set()
            visited_edges[key].add((source,action,target))
    
    
    for k,v in visited_edges.items():
        if len(v)>1:
            for edge in v:
#                 print(edge)
                duplicate_edges.append(edge)

    return duplicate_edges

```

```python
# note book input
map_dir='/share/data/mei-work/kangrui/github/mango/data/'
```

```python
file_names = os.listdir(map_dir)
dup_list=[]
for game_name in file_names:
    print(f'checking game {game_name}.....')
    if game_name in ["acorncourt","weapon","theatre"]:
        continue
    G_eval,G,actions,locations,all2all,all_pairs,walkthrough=get_game_info_with_G_eval(map_dir,game_name)
    if check_same_action_different_target_graph(G_eval):
        dup_list.append(game_name)
```

```python
for game_name in dup_list:
    G_eval,G,actions,locations,all2all,all_pairs,walkthrough=get_game_info_with_G_eval(map_dir,game_name)
    duplicate_edges=find_duplicate_edges(G_eval)
    print(f'{game_name}:')
    for edge in duplicate_edges:
        print(edge)
    print('---------------------------------------------------')

    
```
